## Prepare a list of offices sorted by country, state, city

```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
run("SELECT country,state,city FROM offices order by country,state,city")
```

## How many employees are there in the company?


```python
run("SELECT COUNT(employeeNumber) FROM employees")
```

## What is the total of payments received?


```python
run("SELECT SUM(amount) AS TotalPayment FROM payments")
```

## List the product lines that contain 'Cars'.


```python
run("""
SELECT 
    productLine
FROM
    productlines
WHERE
    productLine LIKE '%cars%'
""")
```

## Report total payments for October 28, 2004.


```python
run("SELECT amount,paymentDate FROM payments WHERE paymentDate='2004-10-28'")
```

## Report those payments greater than 100,000.


```python
run("SELECT* FROM payments WHERE amount>100000 ORDER BY amount")
```

## List the products in each product line.


```python
run("SELECT productLine,productName FROM products ORDER BY productLine")
```

## How many products in each product line?


```python
run("SELECT productLine, COUNT(productLine) As product FROM products GROUP BY productLine")
```

## What is the minimum payment received?


```python
run("SELECT MIN(amount) AS HargaTermurah FROM payments")
```

## List all payments greater than twice the average payment.


```python
run("SELECT AVG(amount)*2 FROM payments")
```
