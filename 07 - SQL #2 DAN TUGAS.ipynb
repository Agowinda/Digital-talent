```python
#mengetahui customer yang tidak memesan atau NULL pada orderNumber atau statusnya
run("""
SELECT 
    c.customerNumber, 
    c.customerName, 
    orderNumber, 
    o.status
FROM
    customers c
        LEFT JOIN
    orders o ON c.customerNumber = o.customerNumber
WHERE
    orderNumber IS NULL
""")
```

# 3. RIGHT JOIN

<img src="http://www.mysqltutorial.org/wp-content/uploads/2017/07/MySQL-Join-right-join-illustration.png" />

```python
run("""
SELECT 
    concat(e.firstName,' ', e.lastName) salesman, 
    e.jobTitle, 
    customerName
FROM
    employees e
        RIGHT JOIN
    customers c ON e.employeeNumber = c.salesRepEmployeeNumber
        AND e.jobTitle = 'Sales Rep'
ORDER BY customerName
""")
```

```python
#menari atasan bawahan
run("""
SELECT 
    CONCAT(m.lastname, ', ', m.firstname) AS 'Manager',
    CONCAT(e.lastname, ', ', e.firstname) AS 'Direct report'
FROM
    employees e
        INNER JOIN
    employees m ON m.employeeNumber = e.reportsto
ORDER BY manager
""")
```

```python
#jika NULL berarti dia top manager
run("""
SELECT 
    IFNULL(CONCAT(m.lastname, ', ', m.firstname),
            'Top Manager') AS 'Manager',
    CONCAT(e.lastname, ', ', e.firstname) AS 'Direct report'
FROM
    employees e
        LEFT JOIN
    employees m ON m.employeeNumber = e.reportsto
ORDER BY manager DESC
""")
```

```python
#customer name> customer name artinya dibandingkan jumlah karakter per alfabetnya
run("""
SELECT 
    c1.city, c1.customerName, c2.customerName
FROM
    customers c1
        INNER JOIN
    customers c2 ON c1.city = c2.city
        AND c1.customername > c2.customerName
ORDER BY c1.city
""")
```

```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
#Report total payments for Atelier graphique.
run("""
SELECT
    customerName,
    SUM(amount)
FROM
    payments
        JOIN customers USING (customerNumber)
WHERE customerName='Atelier graphique'
""")
```

```python
run("""
SELECT
    paymentDate,
    SUM(amount)
FROM
    payments
GROUP BY paymentDate
""")
```

```python
#Report the products that have not been sold
run("""
SELECT 
    p.productCode, 
    p.productName, 
    orderNumber
FROM
    products p
        LEFT JOIN
    orderdetails o ON p.productCode = o.productCode
WHERE
    orderNumber IS NULL
""")
```

```python
#List the amount paid by each customer
run("""
SELECT
    customerName,
    SUM(amount)
FROM
    customers
        JOIN payments USING (customerNumber)
GROUP BY customerName
""")
```

```python
#How many orders have been placed by Herkku Gifts?
run("""
SELECT
    customerName,
    COUNT(orderNumber)
FROM
    customers
        JOIN orders USING (customerNumber)
WHERE customerName='Herkku Gifts'
""")
```

```python
#Who are the employees in Boston?
run("""
SELECT
    employeeNumber,
    CONCAT(firstName,' ',lastName) AS employeeName,
    city
FROM employees
        JOIN offices USING (officeCode)
WHERE city='Boston'
""")
```

```python
#Report those payments greater than $100,000. Sort the report so the customer who made the highest payment appears first.
run("""
SELECT
    customerName,
    amount AS payment
FROM
    customers
        JOIN payments USING(customerNumber)
WHERE amount>100000
ORDER BY amount
""")
```

```python
#List the value of 'On Hold' orders.
run("""
SELECT
    SUM(quantityOrdered*priceEach) AS value,
    status
FROM orderdetails
        JOIN orders USING (orderNumber)
WHERE status='On Hold'
""")
```

```python
#Report the number of orders 'On Hold' for each customer
run("""
SELECT
    customerNumber,
    customerName,
    status
FROM
    customers
        JOIN orders USING(customerNumber)
WHERE status='On Hold'
""")
```

```python
#List products sold by order date
run("""
SELECT
    productName,
    orderDate
FROM products
        JOIN orderdetails USING(productCode) 
        JOIN orders USING(orderNumber)
ORDER BY orderDate
""")
```

```python
#List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
run("""
SELECT
    orderDate,
    orderNumber,
    productName
FROM
    orders
        JOIN orderdetails USING(orderNumber)
        JOIN products USING (productCode)
WHERE productName='1940 Ford Pickup Truck'
ORDER BY orderDate
""")
```

```python
#List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?
run("""
SELECT
    SUM(quantityOrdered*priceEach) AS value,
    customerName,
    orderNumber
FROM
    customers
        JOIN orders USING(customerNumber)
        JOIN orderdetails USING (orderNumber)
GROUP BY orderNumber
HAVING value>25000
""")
```

```python
#Are there any products that appear on all orders?
run
("""
SELECT
    producCode,
    COUNT(*) AS Many
FORM orderdetails
GROUP BY productCode
HAVING Many=(
SELECT COUNT (*) FORM orders)
""")
    
```

```python
#List the names of products sold at less than 80% of the MSRP.
run("""
SELECT
    productName,
    priceEach,
    MSRP
FROM
    products
        JOIN orderdetails USING(productCode)
WHERE 
    priceEach<(MSRP*0.8)
ORDER BY priceEach
""")
```

```python
#Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)
run("""
SELECT
    productName,
    priceEach,
    buyPrice
FROM
    products t1
        JOIN orderdetails t2 ON t1.productCode=t2.productCode
WHERE priceEach>(2*buyPrice)
ORDER BY priceEach
""")
```

```python
#List the products ordered on a Monday.
run("""
SELECT
    DAYNAME(orderDate),
    productName
FROM
    products t1
        JOIN orderdetails t2 ON t1.productCode=t2.productCode
        JOIN orders t3 ON t2.orderNumber=t3.orderNumber
WHERE 
    DAYNAME(orderDate)='monday'
""")
```

```python
#What is the quantity on hand for products listed on 'On Hold' orders?
run("""
SELECT
    SUM(quantityOrdered)
FROM 
    orders t1
        JOIN orderdetails t2 ON t1.orderNumber=t2.orderNumber
WHERE status='On Hold'
""")
```
