# DATA CURAH HUJAN DI BONJONEGORO TAHUN 2012

```python
import pandas as pd
```

```python
data_ch=pd.read_csv('curahhujanbojonegoro.csv',delimiter=',',encoding="ISO-8859-1")
```

```python
data_ch
```

# DESKRIPSI DATA

Dataset ini berisi mengenai data tinggi dari permukaan laut, hari dan curah hujan di Kabupaten Bojonegoro. Penjelasan mengenai variabel dataset di bawah ini:
   
    tahun : Tahun
    kecamatan : Nama Kecamatan
    nama_pos : Nama Pos Pemantau
    lokasi_ketinggian_pos : Meter (m)
    jumlah_stasiun_hujan : Satuan.
    hari_hujan : Hari
    rata-rata_curah_hujan : mm


# STATISTIK SEDERHANA

```python
# statistik sederhana

total=data_ch.rata_curah_hujan.sum()
print("total rata-rata curah hujan bonjonegoro tahun 2012:",total)
```

```python
minimum=data_ch.rata_curah_hujan.min()
print("rata-rata curah hujan minimum bonjonegoro tahun 2012:",minimum)

print("rata-rata curah hujan minimum bonjonegoro tahun 2012 terjadi pada kecamatan Kepohbaru statiun Tlogorejo")
data_ch[data_ch['nama_pos']=='Tlogorejo']
```

```python
maksimum=data_ch.rata_curah_hujan.max()
print("rata-rata curah hujan maksimum bonjonegoro tahun 2012:",maksimum)

print("rata-rata curah hujan maksimum bonjonegoro tahun 2012 terjadi pada kecamatan Kepohbaru stasiun Simorejo")
data_ch[data_ch['nama_pos']=='Simorejo']
```

```python
rata=data_ch.rata_curah_hujan.mean()
print("rata-rata curah hujan bonjonegoro:",rata)
```

```python
median=data_ch.rata_curah_hujan.median()
print("median rata-rata curah hujan bonjonegoro:",median)
```

```python
modus=data_ch.rata_curah_hujan.mode()
print("modus rata-rata curah hujan bonjonegoro:",modus)
```

```python
#jumlah rata-rata curah hujan per kecamatan di bonjonegoro

data_ch.groupby('kecamatan').sum()
```

```python
#rata-rata curah hujan per kecamatan
data_ch.groupby('kecamatan').mean()
```

```python
#jumlah rata-rata curah hujan per pos stasiun
data_ch.groupby('nama_pos').sum()
```

#  VISUALISASI SEDERHANA

```python
ax=data_ch.groupby('kecamatan').sum()['rata_curah_hujan']
ax=ax.plot(kind='bar',title="Jumlah Rata-rata Curah Hujan tiap Kecamatan di Bonjonegoro Tahun 2012",legend=True,fontsize=12)
ax.set_xlabel("kecamatan",fontsize=12)
ax.set_ylabel("rata_curah_hujan",fontsize=12)
```

```python
ax=data_ch.groupby('kecamatan').mean()['rata_curah_hujan']
ax=ax.plot(kind='bar',title="Rata-rata Curah Hujan tiap Kecamatan di Bonjonegoro Tahun 2012",legend=True,fontsize=12)
ax.set_xlabel("kecamatan",fontsize=12)
ax.set_ylabel("rata_curah_hujan",fontsize=12)
```

# ANALISIS DATA

Rata-rata curah hujan tertinggi pada Kabupaten Bonjonegoro terdapat pada Kecamatan Ngasem, sedangkan rata-rata curah hujan terendah terdapat pada kecamatan Gondang. 
Jumlah rata-rata curah hujan tertinggi terjadi pada kecamatan Kepohbaru, sedangkan jumlah rata-rata curah hujan terndah terjadi pada kecamatan Balen.
